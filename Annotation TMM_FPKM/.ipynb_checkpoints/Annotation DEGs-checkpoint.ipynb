```python
import pandas as pd
```

```python
FPKM = pd.read_csv("/Users/csb/Desktop/PartseasonDE/L1r_L3r_L3up.txt", sep='\t') #Trinity ID와 각 샘플의 TMM normallized FPKM 정보
annotation = pd.read_csv("/Users/csb/Desktop/AnnotFPKM/annotation.txt",sep=' ',header=None) #annotation 된 Trinity ID와 accession number
```

```python
FPKM # 데이터 확인
```

```python
annotation # 데이터 확인
```

```python
FPKM_cutoff1 = FPKM[(FPKM['D3B1'] >= 1) | (FPKM['D3B2'] >= 1) | (FPKM['D3B3'] >= 1) | (FPKM['D3B4'] >= 1) | (FPKM['D3L1'] >= 1) | (FPKM['D3L2'] >= 1) | (FPKM['D3L3'] >= 1) | (FPKM['D3L4'] >= 1) | (FPKM['D3S1'] >= 1) | (FPKM['D4B1'] >= 1) | (FPKM['D4B2'] >= 1) | (FPKM['D4B3'] >= 1) | (FPKM['D4B4'] >= 1) | (FPKM['D4L1'] >= 1) | (FPKM['D4L2'] >= 1) | (FPKM['D4L3'] >= 1) | (FPKM['D4L4'] >= 1) | (FPKM['D4S1'] >= 1)]
FPKM = FPKM_cutoff1
FPKM
```

```python
Name = annotation[0] # 이름만 추출
```

```python
Name # 추출 데이터 확인
```

```python
# FPKM 값 넣을 column 만들어주기
annotation["logFC"]=''
annotation["FDR"]=''
annotation["D3B1"]=''
annotation["D3B2"]=''
annotation["D3B3"]=''
annotation["D3B4"]=''
annotation["D3L1"]=''
annotation["D3L2"]=''
annotation["D3L3"]=''
annotation["D3L4"]=''
annotation["D3S1"]=''
annotation["D4B1"]=''
annotation["D4B2"]=''
annotation["D4B3"]=''
annotation["D4B4"]=''
annotation["D4L1"]=''
annotation["D4L2"]=''
annotation["D4L3"]=''
annotation["D4L4"]=''
annotation["D4S1"]=''
```

```python
annotation # column 만들어졌는지 확인
```

```python
# annotation name이 FPKM name 안에 있으면 그 name과 일치하는 TMM normalized FPKM 값을 annotation column에 넣어줌
a=1 # 인풋값과 아웃풋 값의 줄 수가 같은지 확인용
for i in Name:
    if (FPKM['Name']==i).any():
        annotation.loc[annotation[0] == i, "logFC"] = float(FPKM[FPKM.Name==i]['logFC'])
        annotation.loc[annotation[0] == i, "FDR"] = float(FPKM[FPKM.Name==i]['FDR'])
        annotation.loc[annotation[0] == i, "D3B1"] = float(FPKM[FPKM.Name==i]['D3B1'])
        annotation.loc[annotation[0] == i, "D3B2"] = float(FPKM[FPKM.Name==i]['D3B2'])
        annotation.loc[annotation[0] == i, "D3B3"] = float(FPKM[FPKM.Name==i]['D3B3'])
        annotation.loc[annotation[0] == i, "D3B4"] = float(FPKM[FPKM.Name==i]['D3B4'])
        annotation.loc[annotation[0] == i, "D3L1"] = float(FPKM[FPKM.Name==i]['D3L1'])
        annotation.loc[annotation[0] == i, "D3L2"] = float(FPKM[FPKM.Name==i]['D3L2'])
        annotation.loc[annotation[0] == i, "D3L3"] = float(FPKM[FPKM.Name==i]['D3L3'])
        annotation.loc[annotation[0] == i, "D3L4"] = float(FPKM[FPKM.Name==i]['D3L4'])
        annotation.loc[annotation[0] == i, "D3S1"] = float(FPKM[FPKM.Name==i]['D3S1'])
        annotation.loc[annotation[0] == i, "D4B1"] = float(FPKM[FPKM.Name==i]['D4B1'])
        annotation.loc[annotation[0] == i, "D4B2"] = float(FPKM[FPKM.Name==i]['D4B2'])
        annotation.loc[annotation[0] == i, "D4B3"] = float(FPKM[FPKM.Name==i]['D4B3'])
        annotation.loc[annotation[0] == i, "D4B4"] = float(FPKM[FPKM.Name==i]['D4B4'])
        annotation.loc[annotation[0] == i, "D4L1"] = float(FPKM[FPKM.Name==i]['D4L1'])
        annotation.loc[annotation[0] == i, "D4L2"] = float(FPKM[FPKM.Name==i]['D4L2'])
        annotation.loc[annotation[0] == i, "D4L3"] = float(FPKM[FPKM.Name==i]['D4L3'])
        annotation.loc[annotation[0] == i, "D4L4"] = float(FPKM[FPKM.Name==i]['D4L4'])
        annotation.loc[annotation[0] == i, "D4S1"] = float(FPKM[FPKM.Name==i]['D4S1'])
        print(a)
        a+=1
    else:
        idx = annotation[annotation[0] == i].index
        annotation.drop(idx , inplace=True)
```
