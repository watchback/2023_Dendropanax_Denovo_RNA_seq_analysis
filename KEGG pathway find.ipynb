```python
import os
import pandas as pd
```

```python
directory_path = "/Users/csb/Desktop/Project/황칠 Transcriptome/황칠 compound KEGG/황칠 upDEGs 모든 DAVID Pathway,Gene"
new_directory_path = "/Users/csb/Desktop/Project/황칠 Transcriptome/황칠 compound KEGG/황칠 upDEG 모든 DAVID Pathway csv저장"
pathway_directory_path = "/Users/csb/Desktop/pathway_filtered"
```

```python
pathway_list = [
    "Aminobenzoate degradation",
    "Sesquiterpenoid and triterpenoid biosynthesis",
    "Phenylpropanoid biosynthesis",
    "Flavonoid biosynthesis",
    "Flavone and flavonol biosynthesis",
    "Stilbenoid, diarylheptanoid and gingerol biosynthesis",
    "Degradation of flavonoids",
    "Biosynthesis of various plant secondary metabolites",
    "Biosynthesis of phenylpropanoids",
    "Metabolic pathways",
    "Biosynthesis of secondary metabolites",
    "Microbial metabolism in diverse environments",
    "Degradation of aromatic compounds",
    "AMPK signaling pathway",
    "Longevity regulating pathway",
    "Longevity regulating pathway - worm",
]
```

```python
# xlsx csv로 변경
for filename in os.listdir(directory_path): #특정 디렉토리에 있는 모든 파일에 적용
    if filename.endswith('.xlsx'):
        # 파일의 전체 경로
        file_path = os.path.join(directory_path, filename)

        # 파일 읽기
        excel_file = pd.read_excel(file_path)

        # CSV 파일 이름 (확장자 제외)
        csv_filename = filename.split('.')[0] + '.csv'

        # CSV로 저장
        excel_file.to_csv(os.path.join(new_directory_path, csv_filename), index=False)
```

```python
for filename in os.listdir(new_directory_path): #특정 디렉토리에 있는 모든 파일에 적용
    file_path = os.path.join(new_directory_path, filename)
    sample = pd.read_csv(file_path)
    sample["KEGG_PATHWAY"]
    
    filtered_pathway = pd.DataFrame(columns=['ID','Gene Name','Species','KEGG_PATHWAY'])
    
    pathway_count = []

    for i in pathway_list:
        for j in range(len(sample)):
            if i in sample["KEGG_PATHWAY"][j]:
                row_to_add = sample.loc[j,:]
                filtered_pathway = pd.concat([filtered_pathway, pd.DataFrame([row_to_add])], ignore_index=True)
                pathway_count.append(i)
    k = set(pathway_count)
    
    filtered_pathway = filtered_pathway.drop_duplicates()
    filtered_pathway = filtered_pathway.reset_index(drop=True)
    
    n = len(filtered_pathway)
    
    new_filename = filename.split('.')[0] + f'_(Pathway: {len(k)} Gene: {(n)}).csv'
    print(new_filename)
    filtered_pathway.to_csv(os.path.join(pathway_directory_path, new_filename),index=False)
```
