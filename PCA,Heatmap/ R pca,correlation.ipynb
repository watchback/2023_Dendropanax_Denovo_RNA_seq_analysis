```python
r_all = pd.read_csv('/Users/csb/clean_rpca.csv',sep=',',index_col=0) # 파일 불러오기
r_all['category'] = ['B', 'B', 'B', 'B', 'L', 'L', 'L', 'L', 'S','B', 'B', 'B', 'B', 'L', 'L', 'L', 'L', 'S']
r_log = pd.read_csv('/Users/csb/clean_rlogpca.csv',sep=',',index_col=0) # 파일 불러오기
r_log['category'] = ['B', 'B', 'B', 'B', 'L', 'L', 'L', 'L', 'S','B', 'B', 'B', 'B', 'L', 'L', 'L', 'L', 'S']
r_cutoff = pd.read_csv('/Users/csb/clean_cutoff1pca.csv',sep=',',index_col=0) # 파일 불러오기
r_cutoff['category'] = ['B', 'B', 'B', 'B', 'L', 'L', 'L', 'L', 'S','B', 'B', 'B', 'B', 'L', 'L', 'L', 'L', 'S']
r_log_cutoff = pd.read_csv('/Users/csb/clean_logcutoff1pca.csv',sep=',',index_col=0) # 파일 불러오기
r_log_cutoff['category'] = ['B', 'B', 'B', 'B', 'L', 'L', 'L', 'L', 'S','B', 'B', 'B', 'B', 'L', 'L', 'L', 'L', 'S']
```

```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(7,7))
sns.scatterplot(data=r_all,x='PC1',y='PC2',hue=r_all['category'], s=200) # absolute, All genes
plt.xlabel('PC1 (97.5%)', fontsize=15)
plt.ylabel('PC2 (1.5%)', fontsize=15)
for line in range(0,r_all.shape[0]):
     plt.text(r_all.PC1[line]+0.005, r_all.PC2[line], r_all.index[line][1:], horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.savefig('/Users/csb/Desktop/all.jpg')
plt.show()
```
