```python
import pandas as pd
```

```python
# Annotation transcript 정보가 들어있는 파일을 열어서 서열간 엔터키 삭제 
f = open("/Users/csb/Desktop/NRCDS_blast.fasta.txt", 'r')
lines = f.readlines()
ff = open("/Users/csb/Desktop/test.txt", "w")

for line in lines:
    if line[0] == '>':
        ff.write("\n"+line)
    else:
        ff.write(line.replace("\n",""))
```

```python
f2 = pd.read_csv("/Users/csb/Desktop/test.txt", sep=' ', header=None)
```

```python
f2
```

```python
# 각 행의 글자 수 세기
f2['word count'] = f2[0].str.count('')-1
```

```python
f2
```

```python
f3 = f2.copy()
f3
```

```python
f3["word count"] = f3["word count"].shift(-1)
f3 = f3[:-1]
f3
```

```python
# 짝수행(트리니티 ID, 길이)가 들어있는 행 추출
f4 = f3.iloc[::2,:]
```

```python
f4.iloc[:,0] = f4.iloc[:,0].str.replace('>', '')
```

```python
f4
# 길이 추출 완료
```

```python
# csv 저장
f4.to_csv("NRCDS_blast_length.txt", header=False, sep=' ', index=False)
```

```python
fmt6 = pd.read_csv("/Users/csb/Desktop/output3.txt", sep=' ', header=None)
length = pd.read_csv("NRCDS_blast_length.txt", sep=' ', header=None)
length
```

```python
if fmt6[0].equals(length[0]):
    print(f"두 데이터프레임의 열은 동일합니다.")
else:
    print(f"두 데이터프레임의 열은 다릅니다.")
```

```python
fmt6[1] = length[1]
# 정렬은 이미 되어 있으므로 그대로 넣어줌
fmt6
```

```python
# 길이로 정렬
tt = fmt6.sort_values(by=[1],ascending=False)
tt
```

```python
# Uniprot ID가 다른 것만 남겨둠, 길이 기준 긴 것을 남김
ttt = tt.drop_duplicates([2],keep='first')
ttt
```

```python
ttt.to_csv("annotation.txt",header=None,index=False,sep =' ')
```

```python
is_duplicated = ttt[2].duplicated()

# 중복값이 있는지 확인
if is_duplicated.any():
    print(f"데이터프레임의 열에 중복값이 존재합니다.")
else:
    print(f"데이터프레임의 열에 중복값이 없습니다.")
```
