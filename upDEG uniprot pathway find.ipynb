```python
import os
import pandas as pd

a = pd.read_csv("/Users/csb/Desktop/Project/황칠 Transcriptome/황칠 compound Uniprot/Entry_pathway2.csv")

# 특정 디렉토리 경로
directory_path = "/Users/csb/Desktop/Project/황칠 Transcriptome/up DEGs"
save_path = "/Users/csb/Desktop/Project/황칠 Transcriptome/황칠 compound Uniprot/uniprot_up_pathway"
```

```python
for filename in os.listdir(directory_path): #특정 디렉토리에 있는 모든 파일에 적용
    if filename.endswith(".txt"): #.txt 확장자로 끝나는 파일에 적용
        file_path = os.path.join(directory_path, filename)
        csv_filename = os.path.splitext(filename)[0] + ".csv"  # 확장자 변경
        
        b = pd.read_csv(file_path,header=None)
        merged_df = b[b[0].isin(a['Entry'])]
        merged_df.columns = ['Entry']
        result_df = merged_df.merge(a, on='Entry')
        csv_path = os.path.join(save_path, csv_filename)
        result_df.to_csv(csv_path, index=None)
print("작업 완료~")
```

```python
DEGs_list = [
    "Secondary metabolite biosynthesis",
    "Metabolic intermediate biosynthesis",
    "Secondary metabolite metabolism",
    "Aromatic compound metabolism",
    "Phenylpropanoid metabolism",
    "Chorismate biosynthesis",
    "L-phenylalanine degradation",
    "L-phenylalanine biosynthesis",
    "phenylpropanoid biosynthesis",
    "flavonoid biosynthesis",
    "Flavonoid metabolism",
    "3,4',5-trihydroxystilbene biosynthesis",
    "trans-cinnamate biosynthesis",
    "terpenoid biosynthesis",
    "Isoprenoid biosynthesis",
    "abscisate biosynthesis",
    "Terpene metabolism",
    "lanosterol biosynthesis",
    "Sesquiterpene biosynthesis",
]
```

```python
directory_path = "/Users/csb/Desktop/Project/황칠 Transcriptome/황칠 compound Uniprot/uniprot_up_pathway_count"
Counter = 0
DEG_count = pd.DataFrame(columns=['Name', 'Count','Pathway_Count'])
for filename in os.listdir(directory_path): #특정 디렉토리에 있는 모든 파일에 적용
    if filename.endswith(".csv"): #.txt 확장자로 끝나는 파일에 적용
        file_path = os.path.join(directory_path, filename)
        b = pd.read_csv(file_path)
        counter = 0
        Pathwaycounter = 0
        # 특정 리스트의 각 단어에 대해 데이터 프레임 검색 및 Count 열 값 더하기
        for DEGs in DEGs_list:
            mask = (b['Pathway'].str.contains(DEGs)) & (b['Count'] != 0)
            Pathwaycounter += b.loc[mask, 'Pathway'].str.count(DEGs).sum()
            counter += b.loc[mask, 'Count'].sum()
        DEG_count = DEG_count.append({'Name': filename, 'Count': counter, 'Pathway_Count': Pathwaycounter}, ignore_index=True)
DEG_count
```
