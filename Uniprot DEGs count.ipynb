```python
import os
import pandas as pd
```

```python
folder_path = '/Users/csb/Desktop/Project/황칠 Transcriptome/황칠 compound Uniprot/uniprot_up_pathway/' 
save_path = "/Users/csb/Desktop/Project/황칠 Transcriptome/황칠 compound Uniprot/uniprot_up_pathway_count"
```

```python
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(folder_path, filename)

        # CSV 파일 읽기
        a = pd.read_csv(file_path)
        b = pd.read_csv("/Users/csb/Desktop/Project/황칠 Transcriptome/황칠 compound Uniprot/NR_all_pathway.csv")
        for col in b.columns:
            if b[col].dtype == 'O':
                b[col] = b[col].str.rstrip()

        b['Count'] = 0
        target_column = 'Pathway'
        new_column = 'Pathway2'
        new_new_column = 'Pathway3'
        new_new_new_column = 'Pathway4'

        a = a.applymap(lambda x: str(x).replace('.', '') if pd.notnull(x) else x)

        a[new_column]=a[target_column].str.split('; ').apply(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else '')
        a[new_new_column]=a[target_column].str.split('; ').apply(lambda x: x[1] if isinstance(x, list) and len(x) > 1 else '')
        a[new_new_new_column]=a[target_column].str.split('; ').apply(lambda x: x[2] if isinstance(x, list) and len(x) > 2 else '')
        a[new_new_new_column] = a[new_new_new_column].str.replace(': .*', '', regex=True)
        a = a.drop(a.columns[1], axis=1)
        a = a.rename(columns={'Pathway2': 'Pathway', 'Pathway3': 'Pathway2', 'Pathway4': 'Pathway3'})
        for col in a.columns:
            if a[col].dtype == 'O':
                a[col] = a[col].str.rstrip()
        for col in a:
            b['Count'] += b['Pathway'].apply(lambda x: sum(a[col].apply(lambda y: x in y)))
        csv_path = os.path.join(save_path, filename)
        b.to_csv(csv_path, index=None)
print('작업완료~')
```
